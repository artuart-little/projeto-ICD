```python
links_livros = []
print("Iniciando a busca via API...")

# Vamos buscar 5 páginas (20 livros por página = 100 livros)
for pagina in range(1, 6):
    
    # =========================================================
    # PASSO CRÍTICO: Cole aqui a Request URL que você achou no F12.
    # Substitua o número da página pela variável {pagina}
    # =========================================================
    url_api = f"https://prd-api.skoob.com.br/api/v1/book/top-books?page={pagina}&limit=20&ranking_type=read"
    
    print(f"Buscando página {pagina}...")
    resposta = requests.get(url_api, headers=HEADERS)
    
    if resposta.status_code == 200:
        # Como é uma API, lemos a resposta diretamente como JSON (dicionário Python)
        dados_json = resposta.json()
        
        # =========================================================
        # PASSO CRÍTICO 2: Analise a aba "Preview" no F12 para ver o
        # nome exato da lista que contém os livros. 
        # =========================================================
        
        # Substitua 'items' pela chave correta que aparece no seu F12
        lista_de_livros = dados_json.get('items', []) 
        
        for livro in lista_de_livros:
            id_livro = livro.get('edition_id') 
            
            if id_livro:
                # Construindo a URL exata conforme você mencionou
                link_completo = f"https://www.skoob.com.br/pt/book/{id_livro}"
                
                if link_completo not in links_livros:
                    links_livros.append(link_completo)
                    
        time.sleep(random.uniform(1.5, 3.0)) # Pausa educada
        
    else:
        print(f"Erro na página {pagina}. Status: {resposta.status_code}")

# Garantindo os top 100
links_livros = links_livros[:100]

print(f"\nSucesso! {len(links_livros)} links gerados.")
```

```python
def extrair_dados_livro(url):
    """Acessa a página de um livro específico e extrai suas métricas."""
    # Usamos os mesmos HEADERS (com o seu cookie) da Célula 1
    resposta = requests.get(url, headers=HEADERS)
    
    # Dicionário com as chaves exatas que você pediu para o cabeçalho do CSV
    dados = {
        "Título do Livro": None,
        "Autor": None,
        "Quantidade de Páginas": None,
        "Tempo de Leitura": None,
        "Nota Média": None,
        "Leram": None,
        "Lendo": None,
        "Querem": None,
        "Relendo": None,
        "Abandonos": None,
        "Resenhas": None,
        "Favoritos": None,
        "Avaliaram": None      
    }
    
    if resposta.status_code == 200:
        soup = BeautifulSoup(resposta.text, 'html.parser')
        
        try:
            # =========================================================
            # ATENÇÃO: Substitua as strings 'INSPECIONE_AQUI' pelas 
            # classes ou IDs reais que você achar no HTML da página do livro.
            # =========================================================
            
            # Exemplo (ajuste conforme o HTML real):
            # O .text extrai o texto visível e o .strip() remove espaços em branco sobrando
            tag_titulo = soup.find('h1', class_='font-lato text-lg font-bold leading-6')
            if tag_titulo:
                dados["Título do Livro"] = tag_titulo.text.strip()
            
            tag_autor = soup.find('h2', class_='my mt-2 font-lato text-base font-normal leading-5')
            if tag_autor:
                dados["Autor"] = tag_autor.text.strip()    
            
            tag_paginas = soup('span', class_='font-lato text-2xs text-contrast')[2]
            if tag_paginas:
                dados["Quantidade de Páginas"] = tag_paginas.text.strip()

            tag_tempoLeitura = soup('span', class_='font-lato text-2xs text-contrast')[3]
            if tag_tempoLeitura:
                dados["Tempo de Leitura"] = tag_tempoLeitura.text.strip()  

            tag_Nota = soup.find('span', class_='rounded-lg bg-success px-2 py-1 font-lato text-[24px] font-bold leading-8 text-white min-h-8 w-fit min-w-8')
            if tag_Nota:
                dados["Nota Média"] = tag_Nota.text.strip()   

            tag_Leram = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[0]
            if tag_Leram:
                dados["Leram"] = tag_Leram.text.strip()  

            tag_Lendo = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[1]
            if tag_Lendo:
                dados["Lendo"] = tag_Lendo.text.strip()  

            tag_Querem = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[2]
            if tag_Querem:
                dados["Querem"] = tag_Querem.text.strip()

            tag_Relendo = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[3]
            if tag_Relendo:
                dados["Relendo"] = tag_Relendo.text.strip() 

            tag_Abandonos = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[4]
            if tag_Abandonos:
                dados["Abandonos"] = tag_Abandonos.text.strip()  

            tag_Resenhas = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[5]
            if tag_Resenhas:
                dados["Resenhas"] = tag_Resenhas.text.strip()  
                
            tag_Favoritos = soup('span', class_='ml-2')[0]
            if tag_Favoritos:
                dados["Favoritos"] = tag_Favoritos.text.strip()
            
            tag_Avaliaram = soup('span', class_='ml-2')[2]
            if tag_Avaliaram:
                dados["Avaliaram"] = tag_Avaliaram.text.strip()
            # Repita a estrutura acima para todos os outros campos.
            # Dica: Muitos desses números (Leram, Lendo, Abandonos) costumam font-lato text-[24px] font-bold leading-8 text-primary
            # ficar agrupados em uma mesma <div> ou <ul> de estatísticas.
            
        except Exception as e:
            print(f"Erro ao processar a estrutura HTML de {url}: {e}")
            
    else:
        print(f"Falha ao acessar {url} (Status: {resposta.status_code})")
        
    return dados

# --- ÁREA DE TESTE ---
# Rode esta célula e veja se a extração funciona para o PRIMEIRO livro da lista
# antes de rodar para os 100 livros!
link_teste = links_livros[0]
print(f"Testando extração para: {link_teste}")
resultado_teste = extrair_dados_livro(link_teste)
for chave, valor in resultado_teste.items():
    print(f"{chave}: {valor}")
```
