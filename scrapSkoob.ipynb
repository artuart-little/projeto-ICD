```python
#OBTER LINK DE TODOS OS LIVROS

url_api = "https://www.skoob.com.br/pt/search?tab=book"

# Teste nas 3 primeiras páginas. Altere TOTAL_PAGINAS para obter tudo.
TOTAL_PAGINAS = 3
ARQUIVO_LINKS = "links_todos_livros.txt"

with open(ARQUIVO_LINKS, 'a', encoding='utf-8') as f:
    for pagina in range(1, TOTAL_PAGINAS+1): 
        
        payload = [{"params": {"page": pagina, "limit": 12, "searchType": "book"}}]
        
        try:
            resposta = requests.post(url_api, headers=HEADERS, json=payload, timeout=15)
            
            if resposta.status_code == 200:
                texto_bruto = resposta.text
                livros_encontrados = False
                
                for linha in texto_bruto.split('\n'):
                    if '"total_items"' in linha and '"items"' in linha:
                        inicio_json = linha.find('{')
                        if inicio_json != -1:
                            json_limpo = linha[inicio_json:]
                            dados = json.loads(json_limpo)
                            
                            lista_de_livros = dados.get('items', [])
                            for livro in lista_de_livros:
                                id_livro = livro.get('edition_id')
                                
                                if id_livro:
                                    link_completo = f"https://www.skoob.com.br/pt/book/{id_livro}\n"
                                    f.write(link_completo)
                                    
                            print(f"Sucesso! Página {pagina} processada. {len(lista_de_livros)} links extraídos.")
                            livros_encontrados = True
                            break
                
                if not livros_encontrados:
                    print(f"Página {pagina}: Conexão OK, mas não achou os dados. O servidor pode ter retornado HTML de novo.")
                    
            else:
                print(f"Erro na página {pagina}. Status: {resposta.status_code}")
                
        except Exception as e:
            print(f"Erro de conexão na requisição: {e}")
            
        time.sleep(random.uniform(1.5, 3.0))

print(f"\nTeste finalizado! Verifique o arquivo '{ARQUIVO_LINKS}' no seu diretório.")
```

```python
HEADERS = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36",
        "Cookie": MEU_COOKIE
    }

def extrair_dados_livro(url_livro):
    """Acessa a página de um livro específico e extrai suas métricas."""
    
    dados = {
        "Título do Livro": None,
        "Autor": None,
        "Quantidade de Páginas": None,
        "Tempo de Leitura": None,
        "Nota Média": None,
        "Leram": None,
        "Lendo": None,
        "Querem": None,
        "Relendo": None,
        "Abandonos": None,
        "Resenhas": None,
        "Favoritos": None,
        "Avaliaram": None      
    }

    try: 
        resposta = requests.get(url_livro, headers=HEADERS, timeout=15) 
        
        if resposta.status_code == 200:
            soup = BeautifulSoup(resposta.text, 'html.parser')
            
            try:    # O .text extrai texto e o .strip() remove espaços em branco
                tag_titulo = soup.find('h1', class_='font-lato text-lg font-bold leading-6')
                if tag_titulo:
                    dados["Título do Livro"] = tag_titulo.text.strip()
            except: pass

            try:
                tag_autor = soup.find('h2', class_='my mt-2 font-lato text-base font-normal leading-5')
                if tag_autor:
                    dados["Autor"] = tag_autor.text.strip()    
            except: pass    

            try:
                tag_paginas = soup('span', class_='font-lato text-2xs text-contrast')[2]
                if tag_paginas:
                    dados["Quantidade de Páginas"] = tag_paginas.text.strip()
            except: pass 

            try:
                tag_tempoLeitura = soup('span', class_='font-lato text-2xs text-contrast')[3]
                if tag_tempoLeitura:
                    dados["Tempo de Leitura"] = tag_tempoLeitura.text.strip()  
            except: pass 

            try:
                tag_Nota = soup.find('span', class_='rounded-lg bg-success px-2 py-1 font-lato text-[24px] font-bold leading-8 text-white min-h-8 w-fit min-w-8')
                if tag_Nota:
                    dados["Nota Média"] = tag_Nota.text.strip()   
            except: pass 

            try:
                tag_Leram = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[0]
                if tag_Leram:
                    dados["Leram"] = tag_Leram.text.strip()  
            except: pass 

            try:
                tag_Lendo = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[1]
                if tag_Lendo:
                    dados["Lendo"] = tag_Lendo.text.strip()  
            except: pass 

            try:
                tag_Querem = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[2]
                if tag_Querem:
                    dados["Querem"] = tag_Querem.text.strip()
            except: pass 

            try:
                tag_Relendo = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[3]
                if tag_Relendo:
                    dados["Relendo"] = tag_Relendo.text.strip() 
            except: pass 

            try:
                tag_Abandonos = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[4]
                if tag_Abandonos:
                    dados["Abandonos"] = tag_Abandonos.text.strip()  
            except: pass 
            
            try:
                tag_Resenhas = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')[5]
                if tag_Resenhas:
                    dados["Resenhas"] = tag_Resenhas.text.strip()  
            except: pass 
            
            try:
                tag_Favoritos = soup('span', class_='ml-2')[0]
                if tag_Favoritos:
                    dados["Favoritos"] = tag_Favoritos.text.strip()
            except: pass 
            
            try:
                tag_Avaliaram = soup('span', class_='ml-2')[2]
                if tag_Avaliaram:
                    dados["Avaliaram"] = tag_Avaliaram.text.strip()
            except: pass 
                
        else:
            print(f"Erro {resposta.status_code} ao acessar {url_livro}")
                
    except Exception as e:
        print(f"Falha de conexão com {url_livro}: {e}")
        
    return dados

# Teste com o primeiro link do txt
with open(ARQUIVO_LINKS, 'r', encoding='utf-8') as f:
    primeiro_link = f.readline().strip()

print(f"Testando a extração no link: {primeiro_link}")
resultado_teste = extrair_dados_livro(primeiro_link)

for chave, valor in resultado_teste.items():
    print(f"{chave}: {valor}")
```

```python
import pandas as pd
import os
import time
import random

ARQUIVO_CSV = "skoob_todos_livros.csv"

with open(ARQUIVO_LINKS, 'r', encoding='utf-8') as f:
    todos_links = [linha.strip() for linha in f.readlines() if linha.strip()]

livros_raspados = 0
if os.path.exists(ARQUIVO_CSV):
    try:
        df_existente = pd.read_csv(ARQUIVO_CSV)
        livros_raspados = len(df_existente)
        print(f"Retomando progresso: {livros_raspados} livros já estão salvos no CSV.")
    except Exception as e:
        print("Aviso: O CSV existe, mas está vazio ou ilegível. Começando do zero.")

links_para_processar = todos_links[livros_raspados:]
total_restante = len(links_para_processar)

print(f"Iniciando a extração de {total_restante} livros...\n")

for indice, link in enumerate(links_para_processar):
    print(f"[{indice + 1}/{total_restante}] Extraindo: {link}")
    
    dados_extraidos = extrair_dados_livro(link)
    
    df_linha = pd.DataFrame([dados_extraidos])
    
    precisa_cabecalho = not os.path.exists(ARQUIVO_CSV)
    
    df_linha.to_csv(
        ARQUIVO_CSV, 
        mode='a', 
        header=precisa_cabecalho, 
        index=False, 
        encoding="utf-8-sig"
    )
    
    time.sleep(random.uniform(1.0, 2.5))

print("\nExtração da base de dados finalizada com sucesso!")
```
