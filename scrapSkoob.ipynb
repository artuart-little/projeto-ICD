```python
links_livros = []
print("Iniciando a busca via API...")

# Vamos buscar 5 páginas (20 livros por página = 100 livros)
for pagina in range(1, 6):

    # =========================================================
    # PASSO CRÍTICO: Cole aqui a Request URL que você achou no F12.
    # Substitua o número da página pela variável {pagina}
    # =========================================================
    url_api = f"https://prd-api.skoob.com.br/api/v1/book/top-books?page={pagina}&limit=20&ranking_type=read"

    print(f"Buscando página {pagina}...")
    resposta = requests.get(url_api, headers=HEADERS)

    if resposta.status_code == 200:
        # Como é uma API, lemos a resposta diretamente como JSON (dicionário Python)
        dados_json = resposta.json()

        # =========================================================
        # PASSO CRÍTICO 2: Analise a aba "Preview" no F12 para ver o
        # nome exato da lista que contém os livros.
        # =========================================================

        # Substitua 'items' pela chave correta que aparece no seu F12
        lista_de_livros = dados_json.get('items', [])

        for livro in lista_de_livros:
            id_livro = livro.get('edition_id')

            if id_livro:
                # Construindo a URL exata conforme você mencionou
                link_completo = f"https://www.skoob.com.br/pt/book/{id_livro}"

                if link_completo not in links_livros:
                    links_livros.append(link_completo)

        time.sleep(random.uniform(1.5, 3.0)) # Pausa educada

    else:
        print(f"Erro na página {pagina}. Status: {resposta.status_code}")

# Garantindo os top 100
links_livros = links_livros[:100]

print(f"\nSucesso! {len(links_livros)} links gerados.")
```

```python
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
from bs4 import BeautifulSoup
import csv
import concurrent.futures
import time
import os
import threading # Necessário para proteger a gravação no arquivo

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
}

NOME_ARQUIVO_TXT = "links_todos_livros.txt"
NOME_ARQUIVO_CSV = "dados_livros_skoob.csv"
MAX_WORKERS = 4

# A "trava" para organizar a fila na hora de salvar no arquivo
trava_gravacao = threading.Lock()

def extrair_dados_livro(url, sessao):
    # ... (ESTA FUNÇÃO CONTINUA EXATAMENTE IGUAL AO CÓDIGO ANTERIOR) ...
    dados = {
        "URL": url, "Título do Livro": None, "Autor": None, "Quantidade de Páginas": None,
        "Tempo de Leitura": None, "Nota Média": None, "Leram": None, "Lendo": None,
        "Querem": None, "Relendo": None, "Abandonos": None, "Resenhas": None,
        "Favoritos": None, "Avaliaram": None
    }
    try:
        resposta = sessao.get(url, timeout=15)
        if resposta.status_code == 200:
            soup = BeautifulSoup(resposta.text, 'html.parser')
            def pegar_texto(tag): return tag.text.strip() if tag else None
            def pegar_texto_por_indice(lista_tags, i): return lista_tags[i].text.strip() if len(lista_tags) > i else None

            dados["Título do Livro"] = pegar_texto(soup.find('h1', class_='font-lato text-lg font-bold leading-6'))
            dados["Autor"] = pegar_texto(soup.find('h2', class_='my mt-2 font-lato text-base font-normal leading-5'))
            tags_contraste = soup('span', class_='font-lato text-2xs text-contrast')
            dados["Quantidade de Páginas"] = pegar_texto_por_indice(tags_contraste, 2)
            dados["Tempo de Leitura"] = pegar_texto_por_indice(tags_contraste, 3)
            dados["Nota Média"] = pegar_texto(soup.find('span', class_='rounded-lg bg-success px-2 py-1 font-lato text-[24px] font-bold leading-8 text-white min-h-8 w-fit min-w-8'))
            tags_primarios = soup('span', class_='font-lato text-[24px] font-bold leading-8 text-primary')
            dados["Leram"] = pegar_texto_por_indice(tags_primarios, 0)
            dados["Lendo"] = pegar_texto_por_indice(tags_primarios, 1)
            dados["Querem"] = pegar_texto_por_indice(tags_primarios, 2)
            dados["Relendo"] = pegar_texto_por_indice(tags_primarios, 3)
            dados["Abandonos"] = pegar_texto_por_indice(tags_primarios, 4)
            dados["Resenhas"] = pegar_texto_por_indice(tags_primarios, 5)
            tags_ml2 = soup('span', class_='ml-2')
            dados["Favoritos"] = pegar_texto_por_indice(tags_ml2, 0)
            dados["Avaliaram"] = pegar_texto_por_indice(tags_ml2, 2)
    except Exception:
        pass # Silencia o erro para não poluir a tela, o título ficará None
    return dados


if __name__ == "__main__":
    tempo_inicio = time.time()

    # 1. CARREGAR TODAS AS URLs DO TXT
    todas_urls = []
    try:
        with open(NOME_ARQUIVO_TXT, 'r', encoding='utf-8') as arquivo_txt:
            todas_urls = [linha.strip() for linha in arquivo_txt if linha.strip()]
    except FileNotFoundError:
        print(f"❌ Erro: '{NOME_ARQUIVO_TXT}' não encontrado.")
        exit()

    # 2. O PULO DO GATO: VERIFICAR O QUE JÁ FOI LIDO
    urls_ja_lidas = set()
    arquivo_ja_existe = os.path.exists(NOME_ARQUIVO_CSV)

    if arquivo_ja_existe:
        with open(NOME_ARQUIVO_CSV, 'r', encoding='utf-8-sig') as f:
            leitor = csv.DictReader(f, delimiter=';')
            for linha in leitor:
                # Adiciona a URL lida na nossa memória
                urls_ja_lidas.add(linha['URL'])

    # Filtra a lista mantendo apenas os livros que ainda não foram raspados
    urls_pendentes = [url for url in todas_urls if url not in urls_ja_lidas]

    print(f"📊 Resumo inicial:")
    print(f"  - Total de links no TXT: {len(todas_urls)}")
    print(f"  - Links já salvos no CSV: {len(urls_ja_lidas)}")
    print(f"  - Links restantes para raspar agora: {len(urls_pendentes)}")
    print("-" * 50)

    if not urls_pendentes:
        print("🎉 Todos os livros já foram raspados e salvos! Nada a fazer.")
        exit()

    # Configuração de Sessão e Retentativas (Mantido igual)
    sessao_global = requests.Session()
    sessao_global.headers.update(HEADERS)
    estrategia_retry = Retry(total=4, backoff_factor=1, status_forcelist=[429, 500, 502, 503, 504])
    adaptador = HTTPAdapter(max_retries=estrategia_retry)
    sessao_global.mount("https://", adaptador)

    # 3. ABRIR ARQUIVO PARA GRAVAÇÃO EM TEMPO REAL
    # Se já existir, usa 'a' (append). Se não, usa 'w' (write) para criar do zero.
    modo_abertura = 'a' if arquivo_ja_existe else 'w'

    with open(NOME_ARQUIVO_CSV, modo_abertura, newline='', encoding='utf-8-sig') as arquivo_csv:
        # Definimos os campos manualmente para garantir a ordem correta
        campos_csv = [
            "URL", "Título do Livro", "Autor", "Quantidade de Páginas", "Tempo de Leitura",
            "Nota Média", "Leram", "Lendo", "Querem", "Relendo", "Abandonos", "Resenhas",
            "Favoritos", "Avaliaram"
        ]
        escritor = csv.DictWriter(arquivo_csv, fieldnames=campos_csv, delimiter=';')

        # Só escreve o cabeçalho se o arquivo estiver sendo criado agora
        if modo_abertura == 'w':
            escritor.writeheader()

        print(f"🚀 Iniciando extração dos {len(urls_pendentes)} livros restantes...")

        with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
            futuros = {executor.submit(extrair_dados_livro, url, sessao_global): url for url in urls_pendentes}

            for futuro in concurrent.futures.as_completed(futuros):
                resultado = futuro.result()
                titulo = resultado.get('Título do Livro')

                # Só salvamos no CSV se a raspagem foi um sucesso (tem título válido)
                if titulo:
                    with trava_gravacao: # Garante que só uma conexão escreva no arquivo por vez
                        escritor.writerow(resultado)
                        arquivo_csv.flush() # Força o computador a salvar no HD IMEDIATAMENTE!

                    print(f"✓ Salvo: {titulo[:30]}...")
                else:
                    print(f"❌ Falha ignorada. A URL será tentada na próxima vez que rodar o script.")

    print("-" * 50)
    print(f"🎉 Extração finalizada! Os novos dados foram adicionados ao '{NOME_ARQUIVO_CSV}'.")
```
